```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# --- (a) TAREAS BÁSICAS DE EXPLORACIÓN ---

```python
# -------------- CARGAR EL DATASET --------------
# Lee el dataset desde el archivo CSV y lo carga en un DataFrame
df = pd.read_csv("CO2 Emissions_Canada.csv")

# Muestra las primeras filas del dataset para tener una primera visión
print(df.head())

# -------------- ANÁLISIS DESCRIPTIVO --------------
# Proporciona estadísticas descriptivas del dataset
print(df.describe())
```

```python
# Conteo de las Marcas (Make) de Vehículos
plt.figure(figsize=(15,8))
df['Make'].value_counts().plot(kind='bar')
plt.title("Conteo de Marcas de Vehículos")
plt.ylabel("Número de Vehículos")
plt.xlabel("Marca")
plt.xticks(rotation=90)
plt.show()

# Distribución de las Clases de Vehículos
plt.figure(figsize=(15,8))
df['Vehicle Class'].value_counts().plot(kind='bar')
plt.title("Distribución de Clases de Vehículos")
plt.ylabel("Número de Vehículos")
plt.xlabel("Clase de Vehículo")
plt.xticks(rotation=90)
plt.show()

# Distribución de Tipos de Transmisión
plt.figure(figsize=(10,6))
df['Transmission'].value_counts().plot(kind='bar')
plt.title("Distribución de Tipos de Transmisión")
plt.ylabel("Número de Vehículos")
plt.xlabel("Tipo de Transmisión")
plt.xticks(rotation=45)
plt.show()

#Distribución de Tipos de Combustible
plt.figure(figsize=(10,6))
df['Fuel Type'].value_counts().plot(kind='bar')
plt.title("Distribución de Tipos de Combustible")
plt.ylabel("Número de Vehículos")
plt.xlabel("Tipo de Combustible")
plt.show()
```

# --- (b) LIMPIEZA DEL DATASET ---

```python
# -------------- VERIFICACIÓN DE VALORES VACÍOS --------------
# Verifica si hay valores vacíos en el dataset y muestra la suma por columna
print(df.isnull().sum())

# Elimina las filas con valores vacíos
df.dropna(inplace=True)

# Codificación One-Hot para las columnas categóricas
df_encoded = pd.get_dummies(df, columns=['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type'], drop_first=True)

# -------------- GUARDAR EL DATASET PROCESADO --------------
df.to_csv("processed_data.csv", index=False)
```

## --- (c) DOCUMENTACIÓN ---

### Razones para elegir este dataset:
Escogimos este dataset porque el cambio climático es una preocupación creciente a nivel global. Entender las emisiones de CO2 de los vehículos puede ayudar a identificar y promover prácticas más sostenibles. Además, tiene una combinación de variables categóricas y numéricas que lo hace interesante para el análisis.

### Problema a resolver:
Nuestro objetivo es construir un modelo de regresión que pueda predecir las emisiones de CO2 de un vehículo basado en sus características. Esto podría ser útil para evaluar vehículos futuros o para políticas de regulación.

### Detalles del dataset:
- **Model:** Describe el tipo de tracción del vehículo y su distancia entre ejes.
- **Transmission:** Define si el vehículo es automático, manual, y cuántos engranajes tiene.
- **Fuel type:** Tipo de combustible que utiliza el vehículo.
- **Fuel Consumption:** Consumo de combustible tanto en ciudad como en carretera.
- **CO2 Emissions:** Emisiones de CO2 del vehículo (nuestra variable objetivo).
